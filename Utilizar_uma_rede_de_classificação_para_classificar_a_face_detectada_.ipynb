```


import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
plt.imshow(plt.imread('/content/friends.jpg'))
plt.show()


# Load the pre-trained face detection model (Haar Cascade)
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Load the image
img = cv2.imread('/content/friends.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect faces in the image
faces = face_cascade.detectMultiScale(gray, 1.1, 4)

# Define names for the detected faces
names = ["Unknown", "Katy", "Ross", "Martha", "Michelle", "Mike"]

# Draw rectangles around the detected faces and add names
i = 0
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)

    # Use the names list to label the faces
    if i < len(names):
        cv2.putText(img, names[i], (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)
        i += 1
    else:
        cv2.putText(img, "Unknown", (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

# Display the image with detected faces and names
cv2_imshow(img)
```
